```python
# Exercise
# Encode and decode a string by mixing its characters in odd and even locations.
# sample input: hello
# sample output: hlo + el = hloel

def encrypt(text):
# input message
#text='hello'
#text='hello, Mike! This is a big secret: we don\'t have class tmr'

    #print('text=\t',text)
    # make sure it has even length
    if len(text) % 2 == 1:
        text += ' ' # add a space to avoid odd length.


    # encryption/encoding
    #print('------- encryption ---------')
    #get characters at even and odd locations
    even=text[::2]
    #print('even=\t',even) # \t for tab
    odd=text[1::2]
    #print('odd=\t',odd)

    output=even+odd
    #print('output=\t',output)
    return output


def decrypt(output):
    #print('------- decyption ---------')

    length=len(output)
    center=length//2  # // for interger result, this require length to be even
    #print(length,center)
    left = output[:center]
    #print('left=\t',left)
    right = output[center:]
    #print('right=\t',right)

    # how to combine left and right together and decode the message?
    # It is easy with a for loop
    decoded=''
    for i in range(center):
        decoded += (left[i])
        decoded += (right[i])
    #print('decoded=',decoded)
    return decoded
def simulate(text):
    #message='tmr is Saturday!'
    output = encrypt(text)
    decoded = decrypt(output)
    print(output,'(', decoded,')')
```

```python
message='tmr is Saturday!'
output = encrypt(message)
decoded = decrypt(output)

simulate('tmr is Saturday!')

simulate('There is fire on the mountain, which doesn\'t smell good!')

```
