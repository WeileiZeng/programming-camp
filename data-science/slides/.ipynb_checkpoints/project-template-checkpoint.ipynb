```python
# import packages as needed
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
## Data

# example: read a file and show detail
tw = pd.read_csv('../sample-code/data/covid-term/2020-05-03_top1000terms.csv', header=None,names=['term','frequency'])
print(tw.head())

#get entry for ketword 'covid'
a=tw[tw['term']=='covid']
print('a=\n',a)
b=a.set_index(pd.Index(range(a.shape[0])))  #reset the index
print('b=\n',b)
b.loc[0,'frequency']
print('b=\n',b.loc[0]['frequency'])
```

```python
# use it as a function
# read all files and plot some information
def term_plot(keywords):
    t_range=pd.date_range('2020-03-22','2020-07-06')
#t_range
#keywords=['covid']
#keywords=['china']

#keywords=['covid','coronavirus','trump','google','china','wuhan','virus']
#keywords=['trump','google','china','wuhan','virus','doctor','cuomo']

    df = DataFrame(index=t_range)             #a DataFrame to save selected data
    folder='../sample-code/data/covid-term/'  #folder with the data files

    #loop through each day
    for d in t_range:
        filename=folder + str(d)[:10] + '_top1000terms.csv'
        #print('filename --> ',filename)
        tw = pd.read_csv(filename, header=None, names=['term','frequency'])
    
        for key in keywords:
            tw_short=tw[tw['term']==key]
            if tw_short.shape[0]==0:   #check if the key exists or not
                f=0
            else:                      #sometimes there are two duplicates for the entry, not sure why
                tw_short2=tw_short.set_index(pd.Index(range(tw_short.shape[0])))        
                f=tw_short2.loc[0,'frequency']
            df.loc[d,key]=f
    df.plot(figsize=(12,5))
    return df
#keywords=['covid','coronavirus','trump','google','china','wuhan','virus']
#keywords=['happy','angry','mad','sad','insane','crazy']
keywords=['happy','sad','crazy']
df=term_plot(keywords)
```

## Analysis
- 'happy' refers to degree of happiness, 'sad' refer to the degree of sadness
- The level of happiness and sadness are in the same level. Possible reasons are
  - The big events in public are usually related to the welfare of people. For example, when a goverment policy is made, the benefited party will be happy, but the other not. This will produce great excitment and angry at the same time.
  - The US population is so huge, as well as the uses on Twitter. It is common to have various events happen in every corner of the world and induce different emotions. This may just be an overall statistic result that the population always have this amount of happy and sad feeling.
- It looks like there is a correlation between the keywords 'happy' and 'sad'. That is, a peak of 'happy' are related to a peak of 'sad; also, a valley of 'happy' occurs with a valley of 'sad'. These could only happen when the two terms are related to the same events.
