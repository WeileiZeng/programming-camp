```python
# read data from yahoo finance
alibaba = pdr.get_data_yahoo("BABA")
#alibaba.head()
msft = pdr.get_data_yahoo("MSFT")
#msft.head()
zoom = pdr.get_data_yahoo("ZM")
```

```python
# plot and choose data after 2020-01-01
#plt.plot(alibaba['Close'])
plt.figure(figsize=(12,5))
#plt.plot(alibaba[alibaba.index>'2020-01-01']['Close'])
plt.plot(alibaba[alibaba.index>'2020-04-01']['Close'], label='alibaba Close')
plt.legend()
```

```python
# plot multiple data in one plot
plt.figure(figsize=(12,5))
plt.plot(alibaba[alibaba.index>'2020-01-01']['Close'], label='alibaba Close')
plt.plot(msft[msft.index>'2020-01-01']['Close'], label='microsoft Close')
plt.plot(zoom[zoom.index>'2020-01-01']['Close'], label='zoom Close')
plt.legend()
```

## Analysis
- Alibaba and Microsoft have similar trends. They both decrease in March when COVID occur in US, and then recover in April
  - reasons: they are both tech companies
  - How to check it: one can plot more tech companies
- Zoom increases all the time
  - Reason: it is about online meeting. People use it during quranteene.
  - How to check it: plot more such companies.
