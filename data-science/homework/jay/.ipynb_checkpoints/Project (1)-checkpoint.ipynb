## Introduction
- background: There is a Covid-19 Outbreak.
- goal: How stocks are affected.
- data: Yahoo! Finance.
- method: I will use the data to analyze graphs and compare different companies.

```python
import pandas_datareader as pdr
```

```python
amazon = pdr.get_data_yahoo("AMZN")
#alibaba.head()
costco = pdr.get_data_yahoo("COST")
#msft.head()
homedepot = pdr.get_data_yahoo("HD")
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,5))
plt.plot(amazon[amazon.index>'2020-01-01']['Close'], label='amazon Close')
plt.plot(costco[costco.index>'2020-01-01']['Close'], label='costco Close')
plt.plot(homedepot[homedepot.index>'2020-01-01']['Close'], label='homedepot Close')
plt.legend()
```
