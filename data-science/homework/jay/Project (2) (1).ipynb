## Introduction
- background: There is a Covid-19 Outbreak.
- goal: How stocks are affected.
- data: Yahoo! Finance.
- method: I will use the data to analyze graphs and compare different companies.

```python
import pandas_datareader as pdr
```

```python
amazon = pdr.get_data_yahoo("AMZN")
costco = pdr.get_data_yahoo("COST")
homedepot = pdr.get_data_yahoo("HD")
walmart = pdr.get_data_yahoo("WMT")
bestbuy = pdr.get_data_yahoo("BBY")
ebay = pdr.get_data_yahoo("EBAY")
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,5))
plt.plot(amazon[amazon.index>'2020-01-01']['Close'], label='Amazon')
plt.legend()
```

```python
plt.figure(figsize=(12,5))
plt.plot(costco[costco.index>'2020-01-01']['Close'], label='Costco')
plt.plot(homedepot[homedepot.index>'2020-01-01']['Close'], label='Home Depot')
plt.plot(walmart[walmart.index>'2020-01-01']['Close'], label='Walmart')
plt.plot(bestbuy[bestbuy.index>'2020-01-01']['Close'], label='Best Buy')
plt.plot(ebay[ebay.index>'2020-01-01']['Close'], label='eBay')
plt.legend()
```

## Analysis
- All 6 of these companies's stocks have dropped or varied after March, where Covid-19 first became serious in the US. 
- As time went on, the stock values started to rise back and beyond their previous value.
- Amazon varied the most due to the increasing amount of online purchases during quarantine.
